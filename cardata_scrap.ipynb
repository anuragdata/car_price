```python
import selenium
import pandas as pd
import time

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

# importing regex
import re
```

```python
driver=webdriver.Chrome('chromedriver.exe') 
time.sleep(3)

# Opening the homepage of Amazon.in
url = "https://www.cars24.com/buy-used-car?sort=P&search=SUV&storeCityId=8597&pinId=110001"
driver.get(url)
```

```python
product_urls = []
url=driver.find_elements_by_xpath('//a[@class="_9Ue0B"]')
for i in url:
    product_urls.append(i.get_attribute("href"))
    
```

```python
product_urls
```

```python
models=[]
km=[]
location=[]
year=[]
fuel=[]
mode=[]
variant=[]
price=[]


 
for url in product_urls:
    driver.get(url)
    time.sleep(2)
    
        # Extracting models from xpath
    try:
        model = driver.find_element_by_xpath('/html/body/div[1]/div[2]/div[1]/div/div[2]/div/div/div[1]/div[1]/h2')      
        models.append(model.text)
    except NoSuchElementException:
        models.append('-')    
        
        
    time.sleep(2)
    
    # Extracting Brand from xpath
    try:
        driven = driver.find_element_by_xpath('/html/body/div[1]/div[2]/div[1]/div/div[2]/div/div/div[1]/div[1]/ul/li[1]')      
        km.append(driven.text)
    except NoSuchElementException:
        km.append('-')   

        
        
    time.sleep(2)
    
    # Extracting LOCATION from xpath
    try:
        loc = driver.find_element_by_xpath('//p[@class="_2NHUv"]/span')      
        location.append(loc.text)
    except NoSuchElementException:
        location.append('-')   
        
        
        
    
    time.sleep(2)
    
    # Extracting Brand from xpath
    try:
        yr = driver.find_element_by_xpath('/html/body/div[1]/div[2]/div[3]/div/div/div[1]/div/ul/li[8]/strong')      
        year.append(yr.text)
    except NoSuchElementException:
        year.append('-')    
        
        
   
    time.sleep(2)
    
    # Extracting fuel from xpath
    try:
        fu = driver.find_element_by_xpath('/html/body/div[1]/div[2]/div[1]/div/div[2]/div/div/div[1]/div[1]/ul/li[3]')      
        fuel.append(fu.text)
    except NoSuchElementException:
        fuel.append('-')   
        
        
        
    
    time.sleep(2)
    
    # Extracting mode from xpath
    try:
        mod = driver.find_element_by_xpath('//span[@class="_18vo2"]/small')      
        mode.append(mod.text)
    except NoSuchElementException:
        mode.append('-') 
        
        
    time.sleep(2)
    
    # Extracting Variant from xpath
    try:
        var = driver.find_element_by_xpath('//span[@class="_18vo2"]')      
        variant.append(var.text)
    except NoSuchElementException:
        variant.append('-')
        
        
        
    
    time.sleep(2)
    
    # Extracting Brand from xpath
    try:
        cost = driver.find_element_by_xpath('//strong[@class="_2yYvS"]')      
        price.append(cost.text)
    except NoSuchElementException:
        price.append('-')
        
        
print(len(price),len(variant),len(mode),len(fuel),len(year),len(location),len(models),len(km))
    
    

```
